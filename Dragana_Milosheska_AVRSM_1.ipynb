```python
import cv2
import numpy as np
from easyocr import Reader

reader = Reader(['en'])

# Function to find tables
def find_tables(bboxes, object_classes, confidence_level):
    """
    Identifies table bounding boxes.
    Args:
        bboxes: List of bounding boxes.
        object_classes: Class IDs of detected objects.
        confidence_level: Confidence levels for each detection.
    Returns:
        A list of table bounding boxes and their confidence levels.
    """
    tables = []
    table_conf = []
    for i, bbox in enumerate(bboxes):
        if object_classes[i] < 2:  # Class 0 or 1 corresponds to tables
            tables.append(bbox)
            table_conf.append(confidence_level[i])
    return tables, table_conf

# Function to find cells within a table bounding box
def find_cells(tabela_bbox, bboxes, object_classes, confidence_level):
    """
    Identifies cells within a specific table bounding box.
    Args:
        tabela_bbox: Bounding box of the table.
        bboxes: List of bounding boxes.
        object_classes: Class IDs of detected objects.
        confidence_level: Confidence levels for each detection.
    Returns:
        A list of cell bounding boxes and their confidence levels.
    """
    cells = []
    cell_conf = []
    for i, bbox in enumerate(bboxes):
        if object_classes[i] >= 2:  # Class 2 or higher corresponds to cells
            center_x = (bbox[0] + bbox[2]) / 2  # Center X
            center_y = (bbox[1] + bbox[3]) / 2  # Center Y
            if (
                tabela_bbox[0] < center_x < tabela_bbox[2] and
                tabela_bbox[1] < center_y < tabela_bbox[3]
            ):
                cells.append(bbox)
                cell_conf.append(confidence_level[i])
    return cells, cell_conf

# Function to group cells into columns
def find_columns(table, cells):
    """
    Groups cells into columns based on their x-center coordinates.
    Args:
        table: Bounding box of the table.
        cells: List of cell bounding boxes for the table.
    Returns:
        A list of columns, where each column contains a list of cell bounding boxes.
    """
    # Calculate the center x-coordinates of each cell
    cell_centers = [(i, (bbox[0] + bbox[2]) / 2) for i, bbox in enumerate(cells)]

    # Sort cells by their x-center values
    cell_centers.sort(key=lambda x: x[1])  # Sort by x-center coordinate

    # Cluster cells into columns
    columns = []
    current_column = []
    threshold = 20  # Adjust for your data's spacing between columns

    for idx, center_x in cell_centers:
        if not current_column:
            current_column.append(cells[idx])
        else:
            last_center_x = (current_column[-1][0] + current_column[-1][2]) / 2
            if abs(center_x - last_center_x) < threshold:
                current_column.append(cells[idx])
            else:
                columns.append(current_column)
                current_column = [cells[idx]]

    if current_column:
        columns.append(current_column)

    return columns

# Function to perform OCR on a cropped region
def ocr_from_bbox(image, bbox):
    """
    Performs OCR on a specific bounding box region.
    Args:
        image: The image in which the bounding box is located.
        bbox: Bounding box specifying the region of interest.
    Returns:
        Extracted text as a string.
    """
    x_min, y_min, x_max, y_max = map(int, bbox)
    cropped = image[y_min:y_max, x_min:x_max]  # Crop the region
    results = reader.readtext(cropped)  # Perform OCR
    return " ".join([text[1] for text in results])

```

```python
# Main processing pipeline
def process_results(model, documents, output_path):
    """
    Processes the results from the model, extracts tables, cells, and OCR text.
    Args:
        model: The object detection model.
        documents: Input documents to process.
        output_path: Directory to save results.
    """
    # Load results from the model
    results = model(documents)  # Returns a list of Results objects

    # Process each result
    for result_idx, result in enumerate(results):
        # Extract bounding boxes, object classes, and confidence levels
        bboxes = result.boxes.xyxy.cpu().numpy()  # Convert to NumPy array
        object_classes = result.boxes.cls.cpu().numpy()  # Convert to NumPy array
        confidence_level = result.boxes.conf.cpu().numpy()  # Convert to NumPy array

        # Find tables and their confidence levels
        tabeli, tabeli_conf = find_tables(bboxes, object_classes, confidence_level)

        # Find cells for each table
        cells_per_table = []
        confidence_cells_per_table = []
        for table in tabeli:
            cells, cells_conf = find_cells(table, bboxes, object_classes, confidence_level)
            cells_per_table.append(cells)
            confidence_cells_per_table.append(cells_conf)

        # Load the original image
        image = cv2.imread(result.path)

        # Process each table
        for table_idx, table in enumerate(tabeli):
            # Get cells and columns for the current table
            cells = cells_per_table[table_idx]
            koloni_po_tabela = find_columns(table, cells)

            # Extract OCR text from each cell
            extracted_data = []
            for column in koloni_po_tabela:
                column_data = []
                for cell_bbox in column:
                    text = ocr_from_bbox(image, cell_bbox)  # Perform OCR on the cell
                    column_data.append(text)
                extracted_data.append(column_data)

            # Display extracted OCR text for the table
            print(f"Extracted Data for Table {table_idx + 1}:")
            for col_idx, column in enumerate(extracted_data):
                print(f"Column {col_idx + 1}: {column}")

        # Optionally display and save results
        result.show()  # Display the image with bounding boxes
        save_path = f"{output_path}/result_{result_idx + 1}.jpg"
        result.save(filename=save_path)
```

```python
import cv2
import numpy as np
from doctr.models import ocr_predictor
from doctr.io import DocumentFile

# Initialize the doctr OCR predictor
ocr_model = ocr_predictor(pretrained=True)

# Function to find tables
def find_tables(bboxes, object_classes, confidence_level):
    """
    Identifies table bounding boxes.
    Args:
        bboxes: List of bounding boxes.
        object_classes: Class IDs of detected objects.
        confidence_level: Confidence levels for each detection.
    Returns:
        A list of table bounding boxes and their confidence levels.
    """
    tables = []
    table_conf = []
    for i, bbox in enumerate(bboxes):
        if object_classes[i] < 2:  # Class 0 or 1 corresponds to tables
            tables.append(bbox)
            table_conf.append(confidence_level[i])
    return tables, table_conf

# Function to find cells within a table bounding box
def find_cells(tabela_bbox, bboxes, object_classes, confidence_level):
    """
    Identifies cells within a specific table bounding box.
    Args:
        tabela_bbox: Bounding box of the table.
        bboxes: List of bounding boxes.
        object_classes: Class IDs of detected objects.
        confidence_level: Confidence levels for each detection.
    Returns:
        A list of cell bounding boxes and their confidence levels.
    """
    cells = []
    cell_conf = []
    for i, bbox in enumerate(bboxes):
        if object_classes[i] >= 2:  # Class 2 or higher corresponds to cells
            center_x = (bbox[0] + bbox[2]) / 2  # Center X
            center_y = (bbox[1] + bbox[3]) / 2  # Center Y
            if (
                tabela_bbox[0] < center_x < tabela_bbox[2] and
                tabela_bbox[1] < center_y < tabela_bbox[3]
            ):
                cells.append(bbox)
                cell_conf.append(confidence_level[i])
    return cells, cell_conf

# Function to group cells into columns
def find_columns(table, cells):
    """
    Groups cells into columns based on their x-center coordinates.
    Args:
        table: Bounding box of the table.
        cells: List of cell bounding boxes for the table.
    Returns:
        A list of columns, where each column contains a list of cell bounding boxes.
    """
    # Calculate the center x-coordinates of each cell
    cell_centers = [(i, (bbox[0] + bbox[2]) / 2) for i, bbox in enumerate(cells)]

    # Sort cells by their x-center values
    cell_centers.sort(key=lambda x: x[1])  # Sort by x-center coordinate

    # Cluster cells into columns
    columns = []
    current_column = []
    threshold = 20  # Adjust for your data's spacing between columns

    for idx, center_x in cell_centers:
        if not current_column:
            current_column.append(cells[idx])
        else:
            last_center_x = (current_column[-1][0] + current_column[-1][2]) / 2
            if abs(center_x - last_center_x) < threshold:
                current_column.append(cells[idx])
            else:
                columns.append(current_column)
                current_column = [cells[idx]]

    if current_column:
        columns.append(current_column)

    return columns
def ocr_from_bbox(image, bbox):
    """
    Performs OCR on a specific bounding box region using doctr.
    Args:
        image: The image in which the bounding box is located.
        bbox: Bounding box specifying the region of interest.
    Returns:
        Extracted text as a string.
    """
    x_min, y_min, x_max, y_max = map(int, bbox)
    cropped = image[y_min:y_max, x_min:x_max]  # Crop the region

    # Validate the cropped region dimensions
    if cropped.size == 0 or cropped.shape[0] == 0 or cropped.shape[1] == 0:
        return ""  # Return empty string for invalid regions

    # Convert the cropped region to RGB (as doctr expects RGB input)
    cropped_rgb = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)

    # Convert RGB array to a format compatible with doctr
    doc = DocumentFile.from_images([cropped_rgb])  # Pass as a list of images
    result = ocr_model(doc)  # Perform OCR

    # Extract text from doctr result
    text = " ".join(
        word.value
        for page in result.pages
        for block in page.blocks
        for line in block.lines
        for word in line.words
    )
    return text
```

```python
def process_results(model, documents, output_path):
    results = model(documents)  # Returns a list of Results objects

    for result_idx, result in enumerate(results):
        bboxes = result.boxes.xyxy.cpu().numpy()
        object_classes = result.boxes.cls.cpu().numpy()
        confidence_level = result.boxes.conf.cpu().numpy()

        tabeli, tabeli_conf = find_tables(bboxes, object_classes, confidence_level)

        cells_per_table = []
        for table in tabeli:
            cells, _ = find_cells(table, bboxes, object_classes, confidence_level)
            cells_per_table.append(cells)

        image = cv2.imread(result.path)
        if image is None:
            print(f"Failed to load image: {result.path}")
            continue

        for table_idx, table in enumerate(tabeli):
            cells = cells_per_table[table_idx]
            koloni_po_tabela = find_columns(table, cells)

            extracted_data = []
            for column in koloni_po_tabela:
                column_data = []
                for cell_bbox in column:
                    try:
                        text = ocr_from_bbox(image, cell_bbox)
                    except Exception as e:
                        print(f"Error during OCR: {e}")
                        text = ""
                    column_data.append(text)
                extracted_data.append(column_data)

            print(f"Extracted Data for Table {table_idx + 1}:")
            for col_idx, column in enumerate(extracted_data):
                print(f"Column {col_idx + 1}: {column}")

        result.show()
        save_path = f"{output_path}/result_{result_idx + 1}.jpg"
        result.save(filename=save_path)
```
